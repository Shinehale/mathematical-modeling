### Introduction to Euler Method
` Euler Method ` is a simple method to calculate the approximate solution to differential equation model
the mathematical expressions are as follow:
$$
\left\{
    \begin{matrix}
        y_{i + 1} = y_{i} + hf(x_i, y_i),  i = 0,1,2, \dots, n - 1 \\
        y_0 = y(x_0)
    \end{matrix}
\right.
$$
Acutally, the problem we need to solve as hw is three variables in differential equation groups.

First of all, we need to import some important libraries to fullfill our need.

```python
import numpy as np
import matplotlib.pyplot as plt
```

### Task 1 
#### Q1
we use the Euler method to calculate the approximate solution, then have the concrete value of the equation groups.

Finally, plot them in the same figure.

```python
def deriY1(x, y):
    return 3 * x + 2 * y
    
def deriX1(x, y):
    return 2 * x + 3 * y;

deltT = 0.25
n = 3
X1 = [1]
Y1 = [0]
for i in range(n):
    X1.append(X1[i] + deltT * deriX1(X1[i], Y1[i]))
    Y1.append(Y1[i] + deltT * deriY1(X1[i], Y1[i]))

deltT = deltT / 2.0
n = n * 2 
X2 = [1]
Y2 = [0]
for i in range(n):
    X2.append(X2[i] + deltT * deriX1(X2[i], Y2[i]))
    Y2.append(Y2[i] + deltT * deriY1(X2[i], Y2[i]))
```

```python
deltT = 0.25
n = 3
n1 = np.array([i * deltT for i in range(n + 1)])
x1 = 0.5 * np.exp(-n1) + 0.5 * np.exp(5 * n1)
y1 = -0.5 * np.exp(-n1) + 0.5 * np.exp(5 * n1)

deltT /= 2.0
n = n * 2
n2 = np.array([i * deltT for i in range(n + 1)])
x2 = 0.5 * np.exp(-n2) + 0.5 * np.exp(5 * n2)
y2 = -0.5 * np.exp(-n2) + 0.5 * np.exp(5 * n2)
```

```python
plt.figure(figsize = (16, 4))

plt.subplot(2, 1, 1)
plt.axis('off')
rowLabels = ['t', 'approximate x', 'concerete x', 'approximate y', 'concerete y']
cellText = [n1, X1, x1, Y1, y1]
table = plt.table(cellText=cellText, rowLabels=rowLabels, loc='center', cellLoc='center',rowLoc='center')
table.auto_set_font_size(False)
table.set_fontsize(8) 
table.scale(1, 1.5) 
# cellText.append(n1, X1, x1, Y1, y1)

plt.subplot(2, 1, 2)
plt.axis('off')
rowLabels = ['t', 'approximate x', 'concerete x', 'approximate y', 'concerete y']
cellText = [n2, X2, x2, Y2, y2]
table = plt.table(cellText=cellText, rowLabels=rowLabels, loc='center', cellLoc='center',rowLoc='center')
table.auto_set_font_size(False)
table.set_fontsize(8) 
table.scale(1, 1.5) 
```

### Task 2
- first of all, we need to calculate the approximate value of the $B, T$
- plot the $B(t)$
- plot the $T(t)$
- for the third question, I do not know what is meaning of phase plane

```python
def deriB(B, T):
    return  B * (10 - B - T)
    
def deriT(B, T):
    return T * (15 - B - 3 * T)

deltT = 0.1
n = 70
B = [5.0]
T = [2.0]
for i in range(n):
    B.append(B[i] + deltT * deriB(B[i], T[i]))
    T.append(T[i] + deltT * deriT(B[i], T[i]))
n = [i * deltT for i in range(n + 1)]
```

```python
plt.figure(figsize = (16, 6))

plt.plot(n, B, label = '$B(t)$')
plt.plot(n, T, label = '$T(t)$')
plt.yticks(np.arange(2, 8, 0.5))
plt.axhline(y = 7.5, color = 'grey', linestyle = 'dashed')
plt.axhline(y = 2.5, color = 'grey', linestyle = 'dashed')
plt.xlabel('$t$')
plt.ylabel('$X(t)$')
plt.title('Changes in the number of perch and trout over time')
plt.legend()
```
